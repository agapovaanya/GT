```python
import xlrd
file='/Users/anna/Downloads/ulr.xlsx'
rb = xlrd.open_workbook(file)
sheet = rb.sheet_by_index(0)
a=[]
for rownum in range(sheet.nrows):
    row = sheet.row_values(rownum)
    for c_el in row:
        a.append(c_el)

links=a
```

```python
name_final=[]
article_final=[]
kroshki_final=[]
price_final=[]
proizv_final=[]

for r in range(len(links)):
    url =links[r]
    r = requests.get(url)
    html_content = r.text
    soup = BeautifulSoup(html_content,'html.parser')
    
    if soup.findAll('h1',class_="b-product__title b-online-edit")==[]:
        name=''
    else:
        name=soup.findAll('h1',class_="b-product__title b-online-edit")[0].text
        
    if soup.findAll('li',class_="b-product-data__item b-product-data__item_type_sku")==[]:
        article=''
    else:
        article=soup.findAll('li',class_="b-product-data__item b-product-data__item_type_sku")[0].text[5:]
        
    if soup.findAll('li',class_="b-path__item")==[]:
        kroshki=''
    else:
        kroshki=[]
        for j in soup.findAll('li',class_="b-path__item"):
            kroshki.append(j.text)
    
    if soup.findAll('span', class_="b-sticky-panel__price")==[]:
        price=''
    else:
        price=soup.findAll('span', class_="b-sticky-panel__price")[0].text
    
    
    if soup.findAll('td',class_="b-product-info__cell")==[]:
        proizv=''
    else:
        a=soup.findAll('td',class_="b-product-info__cell")
        proizv=''
        for i in range(len(a)):
            if a[i].text=='Производитель \xa0':
                proizv=a[i+1].text
    name_final.append(name)
    article_final.append(article)
    kroshki_final.append(kroshki)
    price_final.append(price)
    proizv_final.append(proizv)
```

```python
table = pd.DataFrame(list(zip(article_final, name_final,proizv_final, price_final,kroshki_final,links))) 

table.to_excel('avtokluch-63.xlsx')

```
