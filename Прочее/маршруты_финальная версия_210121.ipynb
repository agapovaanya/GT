```python

#устаналиваем счетчик, который запускает цикл до тех пор пока код не пройдет всех клиентов, 
#то есть когда у каждого клиента норма визитов=0, то цикл останавливается
day=0
while sum(norm_viz_kontr_drop)>0:
    distance_res=[]
    day+=1
    print(day)
    for b in range(len(den_zapreta_drop)):
        #с каждым проходом уменьшаем на 1 "день запрета", так как один проход=один день
        if den_zapreta_drop[b]!=0:
            den_zapreta_drop[b]=den_zapreta_drop[b]-1
    #print(den_zapreta_drop[21])
    #в данном массиве будет записываться долгота точек, входящих в маршрут
    marsh1_lat=[]
    #в данном массиве будет записываться широта точек, входящих в маршрут
    marsh1_lon=[]
    dista=[]
    adr=[]
    kontr=[]
    #записываем в массив первую точку, откуда стартуют менеджеры
    nach_toch_lat=dolgota_1
    nach_toch_lon=shirota_1
    marsh1_lat.append(nach_toch_lat)
    marsh1_lon.append(nach_toch_lon)
    
    adr.append('Исходная точка')
    kontr.append('Конечная точка')
    dista.append(0)

    vremya_dnya=[]
    vremya_dnya.append(0)
    
    
    
#############################ZDESSS################    
    #записываем обязательные визиты
    for u in range(len(name_kontr_obyaz)):
        
        if den_vizit_obyaz[u]==day:
            #print(den_vizit_obyaz[u],day)
            R = 6373.0
                        
            #всегда берем последнюю точку из массива долготы и широты, чтобы понять какая точка ближайшая к ней (она и будет следующей)
            lat1 = radians(marsh1_lat[-1])
            lon1 = radians(marsh1_lon[-1])
            #берем по очереди все точки (их долготу и широту)
            lat2 = radians(dolg_kontr_obyaz[u])
            lon2 = radians(shir_kontr_obyaz[u])
            
            #расссчитываем расстояние между двумя точками
            dlon = lon2 - lon1
            dlat = lat2 - lat1

            a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            c = 2 * atan2(sqrt(a), sqrt(1 - a))

            distance = R * c
            # записываем полученное значение расстояния в общий массив
            distance_all.append(distance)
    
            adr.append(name_kontr_obyaz[u])
            kontr.append(address_kontr_obyaz[u])
            marsh1_lat.append(dolg_kontr_obyaz[u])
            marsh1_lon.append(shir_kontr_obyaz[u])
            dista.append(distance)
            print(adr)
            
    
#########################################    
    
    
    max_df=[]
    #смотрим какое максимальное количество визитов
    for y in range(len(norm_viz_kontr_drop)):
        if den_zapreta_drop[y]==0:
            max_df.append(norm_viz_kontr_drop[y])
    
    if max_df==[]:
        max_value=0
    else:
        max_value=max(max_df)
    
    #max_value=max(norm_viz_kontr_drop)
    
    
    for i in range(len(norm_viz_kontr_drop)):
        #сначала берем первого попавшегося контрагента у которого максимальное количество визитов, у которого не стоит запрет 
        # и у которого норма визитов больше 0
        if (norm_viz_kontr_drop[i]==max_value and den_zapreta_drop[i]==0 and norm_viz_kontr_drop[i]!=0):
            
            R = 6373.0
                        
            #всегда берем последнюю точку из массива долготы и широты, чтобы понять какая точка ближайшая к ней (она и будет следующей)
            lat1 = radians(marsh1_lat[-1])
            lon1 = radians(marsh1_lon[-1])
            #берем по очереди все точки (их долготу и широту)
            lat2 = radians(dolg_kontr_drop[i])
            lon2 = radians(shir_kontr_drop[i])
            
            #расссчитываем расстояние между двумя точками
            dlon = lon2 - lon1
            dlat = lat2 - lat1

            a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            c = 2 * atan2(sqrt(a), sqrt(1 - a))

            distance = R * c
            # записываем полученное значение расстояния в общий массив
            distance_all.append(distance)
            
            #если расстояние между точка больше 30, то уменьшаем страховое время. 
            #Так как расстояние считается между точками напрямую, то есть погрешность во времени. Из-за этого закладывается страховое время.
            # Но если между точками большое расстояние, то погрешность почти отсутствует и соответсенно ее убираем.
            if distance>30:
                zapas=0.2
            else:
                zapas=zapas_time

            
            # рассчитываем время, которое тратим на одного клиента, это дорога до клиента + время встречи
        
            vremya_na_odnogo_klienta=(distance/skorost)+((distance*zapas)/skorost)+time_vstr
            distance_res.append(distance)
            dista.append(distance)
             # записываем в общий массим затраченное время, чтобы в дальнейшем понимать сколько клиентов мы еще можем посетить 
            vremya_dnya.append(vremya_na_odnogo_klienta)
             #записываем адрес, наименование и координаты клиента, к которому едем
            adr.append(name_kontr_drop[i])
            kontr.append(address_kontr_drop[i])
            marsh1_lat.append(dolg_kontr_drop[i])
            marsh1_lon.append(shir_kontr_drop[i])
            
            #рассчитываем сколько дней после визита нельзя ездить к клиенту, чтобы не получилось так, что съездили сегодня и завтра алгоритм снова предлагает к нему ехать
            if norm_viz_kontr_drop[i]==2:
                den_zapreta_drop[i]=int((rab_dn-kolvo_prosh_dney)/(norm_viz_kontr_drop[i]-koef))
            else:
                den_zapreta_drop[i]=int((rab_dn-kolvo_prosh_dney)/norm_viz_kontr_drop[i])
            norm_viz_kontr_drop[i]=norm_viz_kontr_drop[i]-1

            #после того как нашли такую точку, то останавливаем счетчик
            break
    print(adr)
    fby=0
    glop=0
     # теперь подбираем к данной точке ближайшую, к ней также ближайшую и т.д
    
    #запускаем еще один цикл, который будет считать до тех пор пока не закончится время в дне,
    # glop не равен 4 (останавливает счетчик, если _______)
    # пока не закончились рабочие дни (2 отнимается, так как цикл начинается с 0 значения)
    
    while (sum(vremya_dnya)<=times and glop!=4 and kolvo_prosh_dney<=(rab_dn-2))==True:
        
        
        distance_all=[]
        count_vizit=[]
        
        #рассчитываем расстояние между последней точкой и всеми другими
        for mko in range(len(norm_viz_kontr_drop)):
            
            R = 6373.0
                        
            lat1 = radians(marsh1_lat[-1])
            lon1 = radians(marsh1_lon[-1])
            lat2 = radians(dolg_kontr_drop[mko])
            lon2 = radians(shir_kontr_drop[mko])

            dlon = lon2 - lon1
            dlat = lat2 - lat1

            a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
            c = 2 * atan2(sqrt(a), sqrt(1 - a))

            distance = R * c
            
            
            #если у точки день запрета не равен 0, то записываем ему вместо реального расстояния, расстояние 9999999, чтобы его точно не выбрал алгоритм, как следующую точку 
             # а все расстояния записываем в единый массив
            if den_zapreta_drop[mko]==0:
                distance_all.append(distance)
                count_vizit.append(norm_viz_kontr_drop[mko])
            else:
                distance_all.append(999999999999)
                count_vizit.append(0)


         # из масссива выбираем самое маленькое значение, это и будет следующая точка
        min_distance=min(distance_all)
    
        bui=[]
        uui=[]
        final=[]
        
        # отбираем точки
        for u in range(len(count_vizit)):
             
            if (count_vizit[u]==4 and distance_all[u]<(km+140)):
                bui.append(distance_all[u])
                uui.append(u)
            else:
                if (count_vizit[u]==3 and distance_all[u]<(km+140)):
                    bui.append(distance_all[u])
                    uui.append(u)
                else:
                    if (count_vizit[u]==2 and distance_all[u]<(km+400)):
                        bui.append(distance_all[u])
                        uui.append(u)
                    else:
                        if (count_vizit[u]!=1 and distance_all[u]<km):
                            bui.append(distance_all[u])
                            uui.append(u)
                        else:
                            if (distance_all[u]==min_distance):
                                bui.append(distance_all[u])
                                uui.append(u)
                                

                            
                            
        if len(bui)==0:
            gyu=0
        else:
            min_val=min(bui)
            if min_val==999999999999:

                #lpjl=5
                break
            else:
                
                
                #выбираем точку с минимальным расстоянием и рассчитываем расстояние между ними и дополняем общий массив
                for h in range(len(bui)):

                    if (bui[h]==min_val and norm_viz_kontr_drop[uui[h]]!=0):
                        
                        
                        
                        if distance_all[uui[h]]>diss:
                            zapas=0.2
                        else:
                            zapas=zapas_time

                        distance=distance_all[uui[h]]
                        vremya_na_odnogo_klienta=(distance/skorost)+((distance*zapas)/skorost)+time_vstr
                        
                        
                        
                        lat111 = radians(dolg_kontr_drop[uui[h]])
                        lon111 = radians(shir_kontr_drop[uui[h]])
                        lat211 = radians(dolgota_1)
                        lon211 = radians(shirota_1)

                        dlon11 = lon211 - lon111
                        dlat11 = lat211 - lat111

                        a11 = sin(dlat11 / 2)**2 + cos(lat111) * cos(lat211) * sin(dlon11 / 2)**2
                        c1111 = 2 * atan2(sqrt(a11), sqrt(1 - a11))

                        distance11 = R * c1111
                        if distance11>diss:
                            zapas1=0.2
                        else:
                            zapas1=zapas_time

                        mops1=(distance11/skorost)+((distance11*zapas1)/skorost)+time_vstr
                        
                        #print(sum(vremya_dnya)+vremya_na_odnogo_klienta+mops1)
                        
                        #если время в дне еще есть, то берем еще одного клиента, если нет то устанавливаем значение glop = 4, чтобы цикл прекратил работу
                        if sum(vremya_dnya)+vremya_na_odnogo_klienta+mops1<times:
                            vremya_dnya.append(vremya_na_odnogo_klienta)
                            dista.append(distance)
                            final.append(uui[h])
                            marsh1_lat.append(dolg_kontr_drop[uui[h]])
                            marsh1_lon.append(shir_kontr_drop[uui[h]])
                            adr.append(name_kontr_drop[uui[h]])
                            kontr.append(address_kontr_drop[uui[h]])
                            if norm_viz_kontr_drop[uui[h]]==2:
                                den_zapreta_drop[uui[h]]=int((rab_dn-kolvo_prosh_dney)/(norm_viz_kontr_drop[uui[h]]-koef))
                            else:
                                den_zapreta_drop[uui[h]]=int((rab_dn-kolvo_prosh_dney)/norm_viz_kontr_drop[uui[h]])

                            
                            norm_viz_kontr_drop[uui[h]]=norm_viz_kontr_drop[uui[h]]-1
                            break
                        else:
                            glop=4
                            break
                            
                            

    #если день закончился, то приписываем конечную точку, она равна изначальной 
    if marsh1_lat[-1]==' ':
        # ничего не проиходит
        mnio=0
    else:
        lat11 = radians(marsh1_lat[-1])
        lon11 = radians(marsh1_lon[-1])
        lat21 = radians(dolgota_1)
        lon21 = radians(shirota_1)

        dlon1 = lon21 - lon11
        dlat1 = lat21 - lat11

        a1 = sin(dlat1 / 2)**2 + cos(lat11) * cos(lat21) * sin(dlon1 / 2)**2
        c111 = 2 * atan2(sqrt(a1), sqrt(1 - a1))

        distance1 = R * c111
        if distance1>diss:
            zapas=0.2
        else:
            zapas=zapas_time
        
        mops=(distance1/skorost)+((distance1*zapas)/skorost)+time_vstr

        marsh1_lat.append(dolgota_1)
        marsh1_lon.append(shirota_1)
        adr.append('Конечная точка')
        kontr.append('Конечная точка')
        dista.append(distance1)
        vremya_dnya.append(mops)
    
    
    lat_fin.append(marsh1_lat)
    lon_fin.append(marsh1_lon)
    kk_fin.append(vremya_dnya)
    dist_fin.append(dista)
    adr_fin.append(adr)
    kontr_fin.append(kontr)
    kolvo_prosh_dney+=1
    
    
    #если рабочие дни закончились, а остались клиенты, у которых остались еще визиты, то ставим у них пометку "не успел"
    if kolvo_prosh_dney>(rab_dn-1):
        for g in range(len(norm_viz_kontr_drop)):
            if norm_viz_kontr_drop[g]!=0:
                #bqwer.append(g)
                norm_viz_kontr_drop[g]=0
                final_name_kontr.append(name_kontr_drop[g])
                final_address_kontr.append(address_kontr_drop[g])
                final_dolg_kontr.append(dolg_kontr_drop[g])
                final_shir_kontr.append(shir_kontr_drop[g])
                final_nomer_dnya.append('Не успел')
                final_rasstoyanie.append('')
                final_vremya.append('')
    
```
